```python
import matplotlib.pyplot as plt
import numpy as np

shows = ['The Crown', 'The Gilded Age', 'The Morning Show', 'Slow Horses', 'Mr. & Mrs. Smith', 'Fallout', '3-Body Problem', 'Shogun']
average_ratings = [the_crown_rating, the_gilded_rating, the_morning_show_rating, slow_horses_rating, mr_mrs_smith_rating, fallout_rating, three_body_rating, shogun_rating]

# Create bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(shows, average_ratings, color=['skyblue', 'coral', 'lightgreen', 'gold', 'plum', 'lightblue', 'orange', 'pink'])

# Customize the plot
plt.xlabel('Shows')
plt.ylabel('Average Rating')
plt.title('Average Ratings of Emmy-Nominated Shows for Outstanding Drama Series')
plt.ylim(0, 10)  # Adjust y-axis limit for better visualization
plt.xticks(rotation=45, ha='right')

# Add rating scores above each bar
for bar in bars:
  yval = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 2), ha='center', va='bottom', fontsize=10)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust the spacing between bars
plt.tight_layout()

# Show the plot
plt.show()
```
