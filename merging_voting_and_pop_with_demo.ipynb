```python
# import files and load in data
# import necessary packages
! pip3 install mapclassify
import geopandas as gpd
import mapclassify
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Load data
md = pd.read_csv('/content/model_data-2.csv') # Load in model_data (county data)
vdf = pd.read_csv('./voting_data_with_population-2.csv') # merged voting and pop df
md.drop(columns=['Unnamed: 0'], inplace=True)
md.head()
```

```python


# Map year ranges to match the format in vdf
year_range_mapping = {
    '2006-2010': '2006-2010',
    '2010-2014': '2010-2014',
    '2012-2016': '2012-2016',
    '2014-2018': '2014-2018',
    '2016-2020': '2016-2020'
}

# Apply the mapping to create a new column 'year range' in year_ranges_df
md['year range'] = md['Data File Year'].map(year_range_mapping)


```

```python
# Define the mapping of individual years to year ranges
year_range_mapping = {
    2000: '2006-2010',
    2001: '2006-2010',
    2002: '2006-2010',
    2003: '2006-2010',
    2004: '2006-2010',
    2005: '2006-2010',
    2006: '2006-2010',
    2007: '2006-2010',
    2008: '2006-2010',
    2009: '2006-2010',
    2010: '2006-2010',
    2011: '2010-2014',
    2012: '2010-2014',
    2013: '2010-2014',
    2014: '2010-2014',
    2015: '2012-2016',
    2016: '2012-2016',
    2017: '2012-2016',
    2018: '2012-2016',
    2019: '2016-2020',
    2020: '2016-2020'
}

# Map individual years to year ranges in vdf
vdf['year range'] = vdf['year'].map(year_range_mapping)

# Merge vdf with year_ranges_df on the common column 'year range'
merged_df = pd.merge(vdf, md, on='year range', how='inner')

# Print the merged dataframe
print(merged_df)

# Save the merged dataframe as a CSV file
merged_df.to_csv("merged_data_with_year_ranges.csv", index=False)
```

```python
# Drop duplicates based on specific columns
merged_df = merged_df.drop_duplicates(subset=['year range', 'county_name'])

# Reset index after dropping duplicates
merged_df.reset_index(drop=True, inplace=True)
```
